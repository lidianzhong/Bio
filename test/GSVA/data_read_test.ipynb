### gene expression data

```python
import pandas as pd

# Read the file
expression_df_raw = pd.read_csv('../../data/GDSC/Cell_line_RMA_proc_basalExp.txt', sep='\t')
expression_df_raw.head()
```

```python
expression_df = expression_df_raw.drop(expression_df_raw.columns[1], axis=1)
expression_df.set_index(expression_df.columns[0], inplace=True)
expression_df = expression_df[expression_df.index.notnull()]
expression_df.head()
```

```python
expression_df.head(5000).to_csv('expression_df.csv', index=True)
```
